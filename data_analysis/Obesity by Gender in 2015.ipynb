# Question 1: What is the obesity rate for females nationally in 2015?

```python
#make national its own data frame named "gender_obesity_in_US_national_2015_df"
gender_obesity_in_US_national_2015_df=obesity_in_US_gender_2015[(obesity_in_US_gender_2015['state_name']=='National')]
gender_obesity_in_US_national_2015_df
```

```python
# create filter named "females_in_2015_national" to focus on female rows
females_in_2015_national=gender_obesity_in_US_national_2015_df[(gender_obesity_in_US_national_2015_df['Gender']=='Female')]
females_in_2015_national
```

```python
# created filter named 'females_in_2015_national_avg' to calculate the average obesity rate of females nationally
females_in_2015_national_avg=females_in_2015_national['gender_pc'].mean()
females_in_2015_national_avg
```

```python
# Use "females_in_2015_national_avg" filter to show the national obesity rate average on chart
females_in_2015_national_avg=females_in_2015_national.groupby('Gender')['gender_pc'].mean()
females_in_2015_national_avg
title1='The national female obesity rate average in 2015'
print("{}\n{}".format(title1, '='*len(title1)))
print(females_in_2015_national_avg.sort_values())

```

### Answer: On a national scale, 29.2% of females were obese in 2015. 

# Question 2: What is the obesity rate for males nationally in 2015?

```python
# create filter named "males_in_2015_national" to focus on males 
males_in_2015_national=gender_obesity_in_US_national_2015_df[(gender_obesity_in_US_national_2015_df['Gender']=='Male')]
males_in_2015_national
```

```python
# create filter named "males_in_2015_national_avg" to calculate the average obesity rate of males nationally
males_in_2015_national_avg=males_in_2015_national['gender_pc'].mean()
males_in_2015_national_avg
```

```python
# Use "males_in_2015_national_avg" filter to calculate the national obesity rate average on chart
males_in_2015_national_avg=males_in_2015_national.groupby('Gender')['gender_pc'].mean()
males_in_2015_national_avg
title1='The national male obesity rate average in 2015'
print("{}\n{}".format(title1, '='*len(title1)))
print(males_in_2015_national_avg.sort_values())

```

### Answer: On a national scale, 35.3% of males were obese in 2015.

# Question 3: How do male and female obesity rates compare in 2015?

```python
# create filer named "list_of_dfs" with male and female national average data frames
# create filter named "combined_df" to combine male and female data frames

list_of_dfs = [males_in_2015_national,
               females_in_2015_national]

combined_df = pd.concat(list_of_dfs)

# plot averages for male and females on a bar chart
combined_df.groupby('Gender')['gender_pc'].mean().plot(kind='bar', color= 'pink')
```

- males are more obese than females in 2015

# **Create a new data frame that only has US states

```python
## Gender
# Create a data frame named "obesity_in_US_gender_2015" that excludes the rows 
# with the value "National", Guam", "Puerto Rico", "Virgin Islands" in the LocationDesc column
obesity_in_US_gender_2015=obesity_in_US_gender_2015[~(obesity_in_US_gender_2015['state_name']=='National')]
obesity_in_US_gender_2015=obesity_in_US_gender_2015[~(obesity_in_US_gender_2015['state_name']=='Guam')]
obesity_in_US_gender_2015=obesity_in_US_gender_2015[~(obesity_in_US_gender_2015['state_name']=='Puerto Rico')]
obesity_in_US_gender_2015=obesity_in_US_gender_2015[~(obesity_in_US_gender_2015['state_name']=='Virgin Islands')]
obesity_in_US_gender_2015['state_name'].unique()
```

# Question 4: What state has the highest female obesity rate? What state has the lowest female obesity rate?

```python
# create filter named "females_in_2015_bystate" on "obesity_in_US_gender_2015" to select only females
females_in_2015_bystate=obesity_in_US_gender_2015[(obesity_in_US_gender_2015['Gender']=='Female')]
females_in_2015_bystate
```

```python
# create filter named "females_in_2015_bystate_avg" to calculate the average obesity rate of females by state
females_in_2015_bystate_avg=females_in_2015_bystate['gender_pc'].mean()
females_in_2015_bystate_avg
```

### Answer: The average obesity rate by state for females is 29.2%

```python
# Use "females_in_2015_bystate_avg" filter to show the state by stae female obesity rate average on chart
females_in_2015_bystate_avg = females_in_2015_bystate.sort_values(['state_name'], ascending = [True])
females_in_2015_bystate_avg=females_in_2015_bystate_avg[['state','state_name','gender_pc','Gender']]
females_in_2015_bystate_avg = females_in_2015_bystate_avg.groupby('state_name')['gender_pc'].mean()
females_in_2015_bystate_avg

```

##### Note: This is a state by state breakdown of the average obesity rate for females in 2015

```python

# 1. Set up the figure to plot bar chart
plt.figure(figsize=(20,4))
plt.subplot

# use the 'females_in_2015_bystate_avg' mean data to plot a bar chart and add labels
females_in_2015_bystate_avg.plot(kind='bar')
plt.xlabel('States')
plt.ylabel('Obesity Rate')
plt.title('Female Obesity Rates by State in 2015')


```

```python
# 1. calculate the mean of the gender_pc column grouping by state 
#    sort it in descending order and name the result "females_in_2015"

# 2. get the top 5 ranked states and name the result "top5_states"

# 3. get the bottom 5 ranked states and name the result "bottom5_states"

females_in_2015=females_in_2015_bystate.groupby('state_name')['gender_pc'].mean().sort_values(ascending=False)
top5_states = females_in_2015.head(5)
bottom5_states = females_in_2015.tail(5)



print("Measuring female obesity rates by state.\n")
print(f"The top 5 states are { ', '.join(top5_states.index)}")
print(f"The bottom 5 states are { ', '.join(bottom5_states.index)}")
females_in_2015.sort_values(ascending=False).to_frame()

```

### Answer: Mississippi had the highest female obesity rate in 2015 at 32.95% and Hawaii had the lowest female obesity rate in 2015 at 23.85%

# Question 5: What state has the highest male obesity rate? What state has the lowest male obesity rate? 

```python
# create filter named "males_in_2015_bystate"
males_in_2015_bystate=obesity_in_US_gender_2015[(obesity_in_US_gender_2015['Gender']=='Male')]
males_in_2015_bystate
```

```python
# create filter named "females_in_2015_bystate_avg" to calculate the average obesity rate of females by state
males_in_2015_bystate_avg=males_in_2015_bystate['gender_pc'].mean()
males_in_2015_bystate_avg
```

### Answer: The average obesity rate by state for males is 35.4%

```python
# Use "males_in_2015_bystate_avg" filter to show the state by state female obesity rate average on chart
males_in_2015_bystate_avg = males_in_2015_bystate.sort_values(['state_name'], ascending = [True])
males_in_2015_bystate_avg=males_in_2015_bystate_avg[['state','state_name','gender_pc','Gender']]
males_in_2015_bystate_avg = males_in_2015_bystate_avg.groupby('state_name')['gender_pc'].mean()
males_in_2015_bystate_avg

```

##### Note: This is a state by state breakdown of the average obesity rate for males in 2015

```python

# 1. Set up the figure to plot bar chart
plt.figure(figsize=(20,4))
plt.subplot

# use the 'males_in_2015_bystate_avg' mean data to plot a bar chart and add labels
males_in_2015_bystate_avg.plot(kind='bar')
plt.xlabel('States')
plt.ylabel('Obesity Rate')
plt.title('Male Obesity Rates by State in 2015')


```

```python
# 1. calculate the mean of the gender_pc column grouping by state 
#    sort it in descending order and name the result "males_in_2015"

# 2. get the top 5 ranked states and name the result "top5_states"

# 3. get the bottom 5 ranked states and name the result "bottom5_states"

males_in_2015=males_in_2015_bystate.groupby('state_name')['gender_pc'].mean().sort_values(ascending=False)
top5_states = males_in_2015.head(5)
bottom5_states = males_in_2015.tail(5)



print("Measuring male obesity rates by state.\n")
print(f"The top 5 states are { ', '.join(top5_states.index)}")
print(f"The bottom 5 states are { ', '.join(bottom5_states.index)}")
males_in_2015.sort_values(ascending=False).to_frame()

```
