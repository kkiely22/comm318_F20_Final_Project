```python
names=restaurants_bystate['name'].unique()
```

```python
restaurant_tomap = pd.read_csv('../data/Clean Data/restaurant_tomap.csv')
```

```python
# create data frame named "name_map_df" to clean up the restaurant names 
name_map_df=pd.DataFrame({'name': names, 'new_name': names})
name_map_df=name_map_df.sort_values('name')

# import "restaurant_name_mapping.csv" data frame call it "name_mapping"
name_mapping = pd.read_csv('../data/Clean Data/restaurant_name_mapping.csv')
```

```python


name_map_df.to_csv('../data/Clean Data/restaurant_name_mapping', index=False)

```

```python
# create filter named "name_map_dict" to make a dictionary 
name_map_dict = dict(zip(name_mapping['name'].values, name_mapping['new_name']))

name_map_dict
```

- these restaurants are messy and need to be cleaned up

```python
# import "restaurant_name_mapping-Copy1.csv" data frame, call it "name_mapping_new"
name_mapping_new = pd.read_csv('../data/Clean Data/restaurant_name_mapping-Copy1.csv')
```

```python
# create filter named "name_map_dict_new" to import the data frame with the new restaurant names
name_map_dict_new = dict(zip(name_mapping_new['name'].values, name_mapping_new['new_name']))

name_map_dict_new
```

- this data frame has all of the corrected restaurant names

```python
# new filter named "restaurants_bystate" to add the new restaurant names to the data frame
restaurants_bystate['new_name']=restaurants_bystate['name'].map(name_map_dict_new)
```

```python
restaurants_bystate
```

- all of the restaurant names match in the "restaurants_bystate" data frame

```python
restaurants_bystate.shape
```

- The new column is "new_name"- this is the column that will be kept. The "name" column will be deleted

```python
restaurants_bystate['new_name'].unique()
```

```python
restaurants_bystate['province'].unique()
```

- need to get rid of LA and Co Spgs

```python
# get rid of the rows containing the strings 'LA' and "Co Spgs"
# define filters
contains_LA = restaurants_bystate['province'].str.contains('LA')
contains_CS = restaurants_bystate['province'].str.contains('Co Spgs')

# select rows
restaurants_bystate = restaurants_bystate[-contains_LA & -contains_CS]
restaurants_bystate['province'].unique()
```

- just the 50 states + DC now

```python
restaurants_bystate.columns
```

- do not need country, keys, postalcode, or websites

```python
# create filter named "cols_to_use" to get rid of unnecessary columns
# create dictionary named "col_map" to rename columns
# create filter "restaurant_tomap" as these will be the columns used to put the restaurants on a US map
cols_to_use = ['address','city', 'latitude', 'longitude', 'province', 'new_name']
col_map = {'address': 'Address',
           'city': 'City',
           'latitude': 'Latitude',
           'longitude': 'Longitude',
           'new_name': 'Restaurant',
           'province': 'State'}
restaurant_tomap=restaurants_bystate[cols_to_use].rename(columns=col_map)
restaurant_tomap
```

- "restaurant_tomap" will be used to make a US map of fast food restaurant locations 

```python
# create filter named "cols_to_use" to get rid of unnecessary columns
# create dictionary named "col_map" to rename columns
# create filter "restaurants_bystate2" as these will be the columns used to explore fast food restaurants
cols_to_use = ['city', 'province', 'new_name']
col_map = {'city': 'City',
           'new_name': 'Restaurant',
           'province': 'State'}
restaurants_bystate2=restaurants_bystate[cols_to_use].rename(columns=col_map)
restaurants_bystate2
```

```python
# save data frame 
restaurants_bystate2.to_csv('../data/Clean Data/restaurants_bystate2.csv')
```
