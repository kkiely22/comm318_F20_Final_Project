# 1. 2014 State by State Obesity Rate

```python
# import data frame
obesity_in_US_gender_2014 = pd.read_csv('../data/Clean Data/obesity_in_US_gender_2014.csv')
```

```python
obesity_in_US_gender_2014
```

```python
#make national its own dataframe named "obesity_in_US_national_2014_df"
obesity_in_US_national_2014_df=obesity_in_US_gender_2014[(obesity_in_US_gender_2014['state_name']=='National')]
obesity_in_US_national_2014_df
```

```python
# save data frame
obesity_in_US_national_2014_df.to_csv('../data/Clean Data/obesity_in_US_national_2014_clean.csv')
```

```python
# Create a filter "national_avg_2014" to find the mean obesity rate nationally
national_avg_2014=obesity_in_US_national_2014_df['gender_pc'].mean()
national_avg_2014
```

```python
# Use "national_avg_2014" filter to show the national obesity rate average on chart
national_avg_2014=obesity_in_US_national_2014_df.groupby('state_name')['gender_pc'].mean()
national_avg_2014
title1='The national obesity rate average in 2014'
print("{}\n{}".format(title1, '='*len(title1)))
print(national_avg_2014.sort_values())
```

#### Answer: The national obesity rate average in 2014 was 32.03%

# 4. Plot the national obesity rates per year

```python
list_of_dfs = [obesity_in_US_national_2014_df,
               obesity_in_US_national_2015_df,
               obesity_in_US_national_2016_df]

combined_df = pd.concat(list_of_dfs)
```

```python
combined_df.groupby('year')['gender_pc'].mean()
```

```python
combined_df.groupby('year')['gender_pc'].mean().plot(kind='bar', color= 'red')
```
