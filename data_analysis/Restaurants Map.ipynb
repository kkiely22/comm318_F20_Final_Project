# Map of Fast Food Restaurants Across the US 

- This map shows the density of fast food restaurants across the US. It represents the fast food restaurants per capita. 
- For example, California has the most fast food restaurants but that is because it is a huge state. This data looks at fast food restaurants per capita. A visualization is helpful because it shows that the smaller states have a lot of restaurants for their size. This will help develop a story about the density and convenience of fast food restaurants

# Setup

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy.random as np
import sys
import matplotlib 
import seaborn as sns
import numpy as np
```

# Import Data

```python
restaurant_tomap = pd.read_csv('../data/Clean Data/restaurant_tomap.csv')
```

# Plot Fast Food Restaurants on Map

```python
restaurant_tomap
```

```python
# use "restaurant_tomap" to delete rows with Puerto Rico and Alaska
restaurant_tomap=restaurant_tomap[~(restaurant_tomap['State']=='PR')]
restaurant_tomap=restaurant_tomap[~(restaurant_tomap['State']=='AK')]
restaurant_tomap
```

```python
# create filter named "geom" to plot the longitude and latitude points from the restaurant_tomap data frame
# create filter named "rests_gdf" to plot points on plot
geom=gpd.points_from_xy(restaurant_tomap['Longitude'], restaurant_tomap['Latitude'] )

rests_gdf = gpd.GeoDataFrame(restaurant_tomap, geometry=geom)
```

```python
rests_gdf.plot()
```

```python
rests_gdf=rests_gdf.set_crs('EPSG:4326')
```

```python
# import US map data frame
us_map = gpd.read_file('../data/Raw Data/us-states.geojson')
```

```python
us_map.crs
```

```python
# plot the US map
us_map.plot()
```

```python
us_map
```

```python
# get rid of Puerto Rico and Alaska rows 
us_map=us_map[~(us_map['name']=='Puerto Rico')]
us_map=us_map[~(us_map['name']=='Alaska')]
```

This new map has all the states except for Alaska and Puerto Rico

```python
# save US map
us_map = us_map.reset_index(drop = True)
```

```python
us_map
```

This data frame shows the geometry points of US states to form the US map

```python
# plot latitide an longitude points of fast food restaurants on to US map
base = us_map.plot(color='white', edgecolor='gray', figsize=(15,15))
rests_gdf[rests_gdf['Longitude']<=-60].plot(ax=base, markersize=5, 
                                            alpha=.8, color='red')
```
